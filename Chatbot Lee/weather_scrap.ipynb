```python
def get_info(page_url):
    page = requests.get(page_url)

    # Parse HTML Code With Beautiful Soup
    soup = BeautifulSoup(page.content, 'html.parser')

    # Find all <table> tags with the specified class
    tables = soup.find_all('table', class_="table table-striped table-condensed table-hover nowrap")

    column_name_list = []

    # Set columns name for the dataframe
    df = pd.DataFrame(columns=['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC'])

    for table in tables:
        for row in table.find_all('tr'):
            row_datas = row.find_all('td')

            if(row_datas != []):
                column_name = row_datas[0].text.strip()
                column_name_list.append(column_name)

                jan = row_datas[1].text.strip()
                feb = row_datas[2].text.strip()
                mar = row_datas[3].text.strip()
                apr = row_datas[4].text.strip()
                may = row_datas[5].text.strip()
                jun = row_datas[6].text.strip()
                jul = row_datas[7].text.strip()
                aug = row_datas[8].text.strip()
                sep = row_datas[9].text.strip()
                oct = row_datas[10].text.strip()
                nov = row_datas[11].text.strip()
                dec = row_datas[12].text.strip()

                df = df.append({'JAN': jan,  'FEB': feb, 'MAR': mar, 'APR': apr, 'MAY': may, 'JUN': jun, 'JUL': jul, 'AUG': aug, 'SEP': sep, 'OCT': oct, 'NOV': nov, 'DEC': dec}, ignore_index=True)

    # Set row index as the types of weather information
    df.index = column_name_list
    return df
```

```python
dataframes = []

# Store data into a list for all the links 
for link in link_list:
    destination_df = get_info(link).to_numpy()
    dataframes.append(destination_df)

# Display dataframes
display(dataframes)
```

```python
def Average(arr_num_list):
    # Compute average value
    return sum(arr_num_list) / len(arr_num_list)
```

```python
def Trim_Array(untrimmed_list):
    # Remove unwanted characters
    removetable = str.maketrans('', '', 'd%°F″h')
    trimmed_weather_list = [s.translate(removetable) for s in untrimmed_list]

    # Convert string array to float array
    new_weather_list = np.array(trimmed_weather_list)
    converted_weather_list = new_weather_list.astype(float)

    return converted_weather_list
```

```python
def Average_weather_info(destination_df, index):
    weather_list = destination_df.iloc[index].values
    # Trim the unwanted characters in the arrays by using Trim_Array() function
    trim_weather_list = Trim_Array(weather_list)

    # Get average value by using the Average() function
    average_weather = Average(trim_weather_list)
    return average_weather
```

```python
def Lowest_weather_info(destination_df, index):
    weather_list = destination_df.iloc[index].values
    # Trim the unwanted characters in the arrays by using Trim_Array() function
    trim_weather_list = Trim_Array(weather_list)

    # Get lowest value
    lowest_weather = min(trim_weather_list)
    return lowest_weather
```

```python
def Highest_weather_info(destination_df, index):
    weather_list = destination_df.iloc[index].values
    # Trim the unwanted characters in the arrays by using Trim_Array() function
    trim_weather_list = Trim_Array(weather_list)

    # Get highest value
    highest_weather = max(trim_weather_list)
    return highest_weather
```

```python
lowest_temp_list = []
lowest_clear_list = []
lowest_precipitation_list = []
lowest_rainfall_list = []
lowest_wind_list = []

avg_temp_list = []
avg_clear_list = []
avg_precipitation_list = []
avg_rainfall_list = []
avg_wind_list = []

highest_temp_list = []
highest_clear_list = []
highest_precipitation_list = []
highest_rainfall_list = []
highest_wind_list = []

for dataframe in dataframes:
    # Convert back the list into dataframes
    cdf = pd.DataFrame(dataframe)

    # Store needed weather information in the lists
    # Lowest value throughout the year
    lowest_temp = Lowest_weather_info(cdf, 2)
    lowest_clear = Lowest_weather_info(cdf,4)
    lowest_precipitation = Lowest_weather_info(cdf,5)
    lowest_rainfall = Lowest_weather_info(cdf,6)
    lowest_wind = Lowest_weather_info(cdf,9)

    lowest_temp_list.append(lowest_temp)
    lowest_clear_list.append(lowest_clear)
    lowest_precipitation_list.append(lowest_precipitation)
    lowest_rainfall_list.append(lowest_rainfall)
    lowest_wind_list.append(lowest_wind)

    # Average value throughout the year
    avg_temp = Average_weather_info(cdf,1)
    avg_clear = Average_weather_info(cdf,4)
    avg_precipitation = Average_weather_info(cdf,5)
    avg_rainfall = Average_weather_info(cdf,6)
    avg_wind = Average_weather_info(cdf,9)

    avg_temp_list.append(avg_temp)
    avg_clear_list.append(avg_clear)
    avg_precipitation_list.append(avg_precipitation)
    avg_rainfall_list.append(avg_rainfall)
    avg_wind_list.append(avg_wind)

    # Highest value throughout the year
    highest_temp = Highest_weather_info(cdf, 0)
    highest_clear = Highest_weather_info(cdf, 4)
    highest_precipitation = Highest_weather_info(cdf, 5)
    highest_rainfall = Highest_weather_info(cdf, 6)
    highest_wind = Highest_weather_info(cdf, 9)

    highest_temp_list.append(highest_temp)
    highest_clear_list.append(highest_clear)
    highest_precipitation_list.append(highest_precipitation)
    highest_rainfall_list.append(highest_rainfall)
    highest_wind_list.append(highest_wind)
```

```python
def Month_info(destination_df, col_name):
    # Convert the data to the lists
    check_month_info = destination_df[col_name].tolist()
    # Trim the unwanted characters in the arrays by using Trim_Array() function
    trimmed_month_info = Trim_Array(check_month_info)

    return trimmed_month_info
```
